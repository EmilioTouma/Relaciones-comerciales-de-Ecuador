#Completeness

No hay missing values:

```python
missing_values_export = data_export.isnull().sum()
completeness_export = (1 - (missing_values_export / len(data_export))) * 100
completeness_export
```

```python
missing_values_import = data_import.isnull().sum()
completeness_import = (1 - (missing_values_import / len(data_import))) * 100
completeness_import
```

Ahora, veamos los valores atípicos para encontrar errores:

```python
least_common_values_export = data_export.apply(lambda x: x.value_counts().idxmin())
least_common_values_export
```

```python
least_common_values_import = data_import.apply(lambda x: x.value_counts().idxmin())
least_common_values_import
```
