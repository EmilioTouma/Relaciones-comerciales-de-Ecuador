# Información interesante

##Exportaciones

```python
destino_exp = data_export.groupby('PAIS_DESTINO').agg({'FOB': ['sum']})
destino_exp_sorted = destino_exp.sort_values(
    [('FOB', 'sum')], ascending=False)[('FOB', 'sum')]
destino_exp_sorted.head(5)

#Colombia, Estados Unidos y Perú son los países donde Ecuador exporta más:
```

```python
distrito_exp = data_export.groupby('DISTRITO').agg({'FOB': ['sum']})
distrito_exp_sorted = distrito_exp.sort_values(
    [('FOB', 'sum')], ascending=False)[('FOB', 'sum')]
distrito_exp_sorted.head(5)

#Tulcan es el distito donde mas exportaciones salen:
```

## Importaciones

```python
IVA = data_import.groupby('DESCRIPCION_ARANCELARIA').agg({'IVA': ['sum']}) #Agrupamos por descripcion arancelaria y la suma se realiza por IVA.
IVA_sorted = IVA.sort_values(
    [('IVA', 'sum')], ascending=False)[('IVA', 'sum')]
IVA_sorted.head(5)
```

```python
#¿Cuáles son los países que más pagan ICE?
ICE = data_import.groupby('DESCRIPCION_ARANCELARIA').agg({'ICE_ADVALOREM': ['sum']})
ICE_sorted = ICE.sort_values(
    [('ICE_ADVALOREM', 'sum')], ascending=False)[('ICE_ADVALOREM', 'sum')]
ICE_sorted.head(5)
```

```python
distrito_import = data_import.groupby('DISTRITO').agg({'FOB': ['sum']})
distrito_import_sorted = distrito_import.sort_values(
    [('FOB', 'sum')], ascending=False)[('FOB', 'sum')]
distrito_import_sorted.head(5)
```
